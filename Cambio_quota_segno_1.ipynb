# INSERIRE LE QUOTE :
# quota_1 = segno 1 quota primo sito
# quota_2 = segno 2 quota primo sito

# quota_3 = segno 1 quota secondo sito
# quota_4 = segno 2 quota secondo sito

# puntata_2 = puntata su segno 2 primo sito

# portafoglio = soldi disponibili totali( puntata_2+puntata successiva )

```python
import pandas as pd
import numpy as np


quota_1 = 1.50
quota_2 = 7

quota_3 = 1.65
quota_4 = 6.50


puntata_2 = 10

puntata_3 = range( 1,1000 )

portafoglio = 80

v1 = []
v2 = []
diff_eff_1 = []
diff_eff_2 = []
q2 = []
q3 = []


for val in puntata_3 :
    if puntata_2+val <= portafoglio :
        diff_eff_1.append( quota_2*puntata_2-puntata_2-val )
        diff_eff_2.append( quota_3*val-puntata_2-val )
        v1.append( puntata_2 )
        v2.append( val )
        q2.append( quota_2 )
        q3.append( quota_3 )
    
```

```python
RISULTATI_RICERCA = pd.DataFrame()

RISULTATI_RICERCA['q2'] = q2
RISULTATI_RICERCA['q3'] = q3
RISULTATI_RICERCA['puntata_2'] = v1
RISULTATI_RICERCA['puntata_3'] = v2
RISULTATI_RICERCA['diff_eff_2'] = diff_eff_1
RISULTATI_RICERCA['diff_eff_3'] = diff_eff_2
RISULTATI_RICERCA['q2'] = q2
RISULTATI_RICERCA['q3'] = q3
RISULTATI_RICERCA
```

```python
over = 0.00

RISULTATI_POSITIVES_DIFFS_over = RISULTATI_RICERCA.loc[( RISULTATI_RICERCA.diff_eff_2>over ) &
                                                              ( RISULTATI_RICERCA.diff_eff_3 >over ) ]

RISULTATI_POSITIVES_DIFFS_over
```

```python
RISULTATI_POSITIVES_DIFFS_over_sorted_diff_1 = RISULTATI_POSITIVES_DIFFS_over.sort_values('diff_eff_2',ascending=False)
RISULTATI_POSITIVES_DIFFS_over_sorted_diff_1
```

```python
RISULTATI_POSITIVES_DIFFS_over_sorted_diff_2 = RISULTATI_POSITIVES_DIFFS_over.sort_values('diff_eff_3',ascending=False)
RISULTATI_POSITIVES_DIFFS_over_sorted_diff_2
```
