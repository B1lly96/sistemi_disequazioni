```python
import pandas as pd
import numpy as np


quota_1 = 1.50
quota_2 = 7.00

puntata_1 = range( 1,1000 )
puntata_2 = range( 1,1000 )


v1 = []
v2 = []
diff_eff_1 = []
diff_eff_2 = []


for v in puntata_1 :
    for val in puntata_2 :
        diff_eff_1.append( quota_1*v-v-val )
        diff_eff_2.append( quota_2*val-v-val )
        v1.append( v )
        v2.append( val )
```

```python
RISULTATI_RICERCA = pd.DataFrame()
RISULTATI_RICERCA['puntata_1'] = v1
RISULTATI_RICERCA['puntata_2'] = v2
RISULTATI_RICERCA['diff_eff_1'] = diff_eff_1
RISULTATI_RICERCA['diff_eff_2'] = diff_eff_2
RISULTATI_RICERCA
```

```python
# BEST diff_eff_1
RISULTATI_RICERCA_diff_1 = RISULTATI_RICERCA.sort_values('diff_eff_1', ascending=False )
RISULTATI_RICERCA_diff_1
```

```python
# BEST diff_eff_2
RISULTATI_RICERCA_diff_2 = RISULTATI_RICERCA.sort_values('diff_eff_2', ascending=False )
RISULTATI_RICERCA_diff_2
```

```python
RISULTATI_POSITIVES_DIFFS = RISULTATI_RICERCA.loc[ ( RISULTATI_RICERCA.diff_eff_1 > 0 )
                                                          & ( RISULTATI_RICERCA.diff_eff_2 > 0 ) ]

RISULTATI_POSITIVES_DIFFS
```

```python
over = 100.00

RISULTATI_POSITIVES_DIFFS_over = RISULTATI_POSITIVES_DIFFS.loc[( RISULTATI_POSITIVES_DIFFS.diff_eff_1>over ) &
                                                              ( RISULTATI_POSITIVES_DIFFS.diff_eff_2 >over ) ]

RISULTATI_POSITIVES_DIFFS_over
```

```python
RISULTATI_POSITIVES_DIFFS_over_sorted_diff_1 = RISULTATI_POSITIVES_DIFFS_over.sort_values('diff_eff_1',ascending=False)
RISULTATI_POSITIVES_DIFFS_over_sorted_diff_1
```

```python
RISULTATI_POSITIVES_DIFFS_over_sorted_diff_2 = RISULTATI_POSITIVES_DIFFS_over.sort_values('diff_eff_2',ascending=False)
RISULTATI_POSITIVES_DIFFS_over_sorted_diff_2
```
