# INSERIRE LE QUOTE :
# quota_1 = segno 1 quota primo sito
# quota_2 = segno 2 quota primo sito

# quota_3 = segno 1 quota secondo sito
# quota_4 = segno 2 quota secondo sito

# puntata_1 = puntata su segno 1 primo sito

# portafoglio = soldi disponibili totali( puntata_1+puntata successiva )

```python
import pandas as pd
import numpy as np


quota_1 = 1.50
quota_2 = 7

quota_3 = 1.65
quota_4 = 6.50


puntata_1 = 20

puntata_4 = range( 1,1000 )

portafoglio = 100


q1 = []
q4 = []
v1 = []
v2 = []
diff_eff_1 = []
diff_eff_2 = []


for val in puntata_4 :
    if puntata_1+val <= portafoglio :
        diff_eff_1.append( quota_1*puntata_1-puntata_1-val )
        diff_eff_2.append( quota_4*val-puntata_1-val )
        v1.append( puntata_1 )
        v2.append( val )
        q1.append( quota_1 )
        q4.append( quota_4 )
```

```python
RISULTATI_RICERCA = pd.DataFrame()

RISULTATI_RICERCA['quota_1'] = q1
RISULTATI_RICERCA['quota_4'] = q4
RISULTATI_RICERCA['puntata_1'] = v1
RISULTATI_RICERCA['puntata_4'] = v2
RISULTATI_RICERCA['diff_eff_1'] = diff_eff_1
RISULTATI_RICERCA['diff_eff_4'] = diff_eff_2
RISULTATI_RICERCA
```

```python
over = 0.00

RISULTATI_POSITIVES_DIFFS_over = RISULTATI_RICERCA.loc[( RISULTATI_RICERCA.diff_eff_1>over ) &
                                                              ( RISULTATI_RICERCA.diff_eff_4 >over ) ]

RISULTATI_POSITIVES_DIFFS_over
```

```python
RISULTATI_POSITIVES_DIFFS_over_sorted_diff_1 = RISULTATI_POSITIVES_DIFFS_over.sort_values('diff_eff_1',ascending=False)
RISULTATI_POSITIVES_DIFFS_over_sorted_diff_1
```

```python
RISULTATI_POSITIVES_DIFFS_over_sorted_diff_2 = RISULTATI_POSITIVES_DIFFS_over.sort_values('diff_eff_4',ascending=False)
RISULTATI_POSITIVES_DIFFS_over_sorted_diff_2
```
