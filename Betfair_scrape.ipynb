```python
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np


url = 'https://www.betfair.it/sport/tennis'

req = urlopen( url )
req
```

```python
bs_betfair = BeautifulSoup( req )
bs_betfair
```

```python
quote = bs_betfair.findAll( "span" )
quote
```

```python
# findAll players( tag='span', class='team-name' )

players = bs_betfair.findAll( "span", { "class":"team-name"})


PLAYERS = []

for v in players :
    PLAYERS.append( v.get_text() )
    
players = []
    
for v in PLAYERS :
    v = v.split("\n")[1]
    players.append( v )
    
numero_players = len( players )
print( players )

```

```python
print( numero_players )
```

```python
base_p1 = list( range(0,numero_players,2) )
base_p2 = list( range(1,numero_players,2) )

print( base_p1 )
print( base_p2 )
```

```python
p1 = [] 
p2 = []

for i,e in enumerate ( base_p1 ) :
    p1.append( players[e] )
print( p1 )

for i,e in enumerate ( base_p2 ) :
    p2.append( players[e] )
print( p2 )
    
```

```python
# tutte le partite sulla pagina


PARTITE = []

for i,v in enumerate ( p1 ) :
    PARTITE.append( v+" - "+p2[i] )
    
PARTITE
```

```python
quote_betfair = bs_betfair.findAll( "span", {"class":re.compile("ui-runner")})
```

```python
QUOTE = []

for e in quote_betfair :
    QUOTE.append( e.get_text() )


quote = []

for v in QUOTE :
    v = v.split("\n")[1]
    quote.append( v )
    
numero_quote = len( quote )
print( quote )
```

```python
q1 = [] 
q2 = []

for i,e in enumerate ( base_p1 ) :
    q1.append( quote[e] )
print( q1 )

for i,e in enumerate ( base_p2 ) :
    q2.append( quote[e] )
print( q2 )
```

```python
# segno 1 per ogni partita
# segno 2 per ogni partita


segno_1 = []
segno_2 = []

for i,v in enumerate ( q1 ) :
    segno_1.append( v )
    
for i,v in enumerate ( q2 ) :
    segno_2.append( v )

    
print( segno_1 )
print( segno_2 )
```

```python
# conversione quote da stringhe a float

segno_1_float = []
segno_2_float = []

for v in segno_1 :
    v = float( v )
    segno_1_float.append( v )
    
for v in segno_2 :
    v = float( v )
    segno_2_float.append( v )
    
print( segno_1_float )
print( segno_2_float )
    
```
