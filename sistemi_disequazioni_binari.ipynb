```python
quota_1 = 1.50
quota_2 = 7.00

puntata_1 = range( 1,1000 )
puntata_2 = range( 1,1000 )

print( "p1  p2  win_1 win_2")
print("_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _")
for v in puntata_1 :
    for val in puntata_2 :
        diff_eff_1 = quota_1*v-v-val
        diff_eff_2 = quota_2*val-v-val
        if ( diff_eff_1>200 ) & ( diff_eff_2>200 ) :
            print( v, val, diff_eff_1, diff_eff_2 ) 
```
